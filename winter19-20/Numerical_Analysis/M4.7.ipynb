```julia
using Printf

maxiter = 10        # maksymalna liczba iteracji w metodzie Newtona
a = Array{Float64,1}([1,2,3,4,5])
n = 4

b = Array{Float64,1}([0,0,0,0,0])
b[n+1] = a[n+1]     # b_n     = a_n
c = Array{Float64,1}([0,0,0,0,0])
c[n+1] = 0.0        # c_n     = 0
c[n]   = a[n+1]     # c_{n-1} = a_n

u,v = 0.0, 0.0

for j = 1:maxiter
    b[n] = a[n] + u*b[n+1]
    for k=n-2:-1:0
        b[k+1] = a[k+1] + u*b[k+2] + v*b[k+3]
        c[k+1] = b[k+2] + u*c[k+2] + v*c[k+3]
    end
    J = c[1]*c[3] - c[2]*c[2]
    u = u + (c[2]*b[2] - c[3]*b[1])/J
    v = v + (c[2]*b[1] - c[1]*b[2])/J
    @show j,u,v,b[1],b[2]
end
```

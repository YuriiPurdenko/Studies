```julia
setprecision(16)
c = [BigFloat(0),BigFloat(0)]
a = 2
while a < 257
    b = (BigFloat(1/2)BigFloat((1+c[a])))
    d = BigFloat(sqrt(b))
    push!(c, d)
    a += 1
end
s = [BigFloat(1),BigFloat(1)]
a = 2
while a < 257
    b = (BigFloat(1/2)BigFloat((1-c[a])))
    d = BigFloat(sqrt(b))
    push!(s, d)
    a += 1
end
p = [BigFloat(2)]
a = 2
while a < 257
    zm = a
    b = 1
    while zm > 1
        b = BigFloat(2*b)
        zm -= 1
    end
    b = BigFloat(s[a]*b)
    push!(p,b)
    print(a)
    print(" P: ")
    print(p[a-1])
    print(" sin: ")
    print(s[a])
    print(" cos:")
    println(c[a-1])
    a += 1
end
```

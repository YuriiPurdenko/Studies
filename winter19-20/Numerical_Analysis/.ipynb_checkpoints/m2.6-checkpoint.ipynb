```julia
c = [BigFloat(0,128),BigFloat(0,128)];
a = 2
while a < 257
    b = (BigFloat(1/2,128)BigFloat((1+c[a]),128))
    d = BigFloat(sqrt(b),128)
    push!(c, d)
    a += 1
end
s = [BigFloat(1,128),BigFloat(1,128)]
a = 2
while a < 257
    b = (BigFloat(1/2,128)BigFloat((1-c[a]),128))
    d = BigFloat(sqrt(b),128)
    push!(s, d)
    a += 1
end
p = [BigFloat(2,128)]
a = 2
while a < 257
    zm = a
    b = 1
    while zm > 1
        b = BigFloat(2*b,128)
        zm -= 1
    end
    b = BigFloat(s[a]*b,128)
    push!(p,b)
    a += 1
end
```

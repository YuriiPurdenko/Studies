```julia
setprecision(128)
sta = -5
fin = 5
n = 40
function func(x)
    return Base.MathConstants.e^(atan(x))
end
```

```julia
function wielomian(start,finish, n)
    s = Float64(start)
    f = Float64(finish)
    h = (f - s)/n
    x = s
    a = [Float64(1),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0)]
    b = [Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0)]
    a[1] = 1
    
    for j = 1:n+1
        for i = 1:n+1
            b[i+1] = a[i]
            a[i] *= x
        end
        for i = 1:n+1
            a[i] += b[i]
        end
        x += h
    end
    print(a)
    print("\n")
    return a
end

function wspol(start, finish, n, k)
    s = Float64(start)
    f = Float64(finish)
    h = (f - s)/n
    x = s
    y = func(x + k * h)
    pos = 0
    odp = 0
    for i = 0 : n
        pos = y
        for j = 0 : n
            if j != i
                pos /= (i - j)*h
            end
        end
        print("\nwtf ")
        print(pos)
        print(" wtf\n")
        odp += pos
    end
    print(odp)
    print(y)
    print("\n")
    print("\n")
    return odp
end
```

```julia
function interpol(start, finish, n)
    c = [Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0),Float64(0)]
    a = wielomian(start, finish, n)
    for i = 0:n
        b = wspol(start, finish, n, i)
        for j = 1:n+1
            c[j] += b * a[j]
        end
    end
    return c
end
            
```
